#Prediksi Terkena Penyakit Paru - Paru

##Penjelasan Dataset

**Sumber Data**: [Kaggle] https://www.kaggle.com/datasets/andot03bsrc/dataset-predic-terkena-penyakit-paruparu/data

**Tujuan Dataset**: Untuk mendiagnosis risiko penyakit yang mungkin terkait dengan kebiasaan merokok, berdasarkan sejumlah atribut individu.

**Atribut dalam Dataset**:

- No: Nomor unik untuk setiap entri (integer).
- Usia: Usia individu (kategori atau ordinal).
- Jenis_Kelamin: Jenis kelamin individu (kategori).
- Merokok: Status merokok individu (kategori).
- Bekerja: Status pekerjaan individu (kategori).
- Rumah_Tangga: Status tanggungan rumah tangga (kategori).
- Aktivitas_Begadang: Frekuensi begadang (kategori).
- Aktivitas_Olahraga: Frekuensi berolahraga (kategori).
- Asuransi: Kepemilikan asuransi kesehatan (kategori).
- Penyakit_Bawaan: Riwayat penyakit bawaan (kategori).

**Target**:
- Hasil: Diagnosis akhir atau risiko kesehatan (kategori).

**Ukuran Dataset**: 30.000 baris dan 11 kolom.

**Tipe Data**:

- Kolom numerik: Hanya kolom No.
- Kolom kategori: 10 kolom lainnya.

##Data Understanding

**Jumlah Data**: 30.000 sampel.

**Distribusi Data**:
- Tidak ada nilai yang hilang di setiap kolom.
- No merupakan identifikasi unik, tidak relevan untuk analisis langsung.
- Semua kolom selain No adalah kategori, sehingga analisis lebih lanjut dapat fokus pada distribusi kategori per kolom.

**Deskripsi Statistik Kolom Numerik**:
- No: Rentang dari 1 hingga 30.000.
- Kolom Kategori: Perlu eksplorasi lebih lanjut untuk mengetahui distribusi setiap kategori.

##Business Understanding

**Tujuan Bisnis**:
- Membantu dalam diagnosis awal risiko kesehatan berdasarkan kebiasaan dan atribut individu.
- Menyediakan alat bantu bagi sistem pendukung keputusan dalam mendeteksi risiko penyakit.

**Manfaat Bisnis**:
- Meningkatkan kesadaran kesehatan masyarakat tentang faktor risiko.
- Memungkinkan intervensi dini terhadap risiko penyakit terkait kebiasaan merokok.
- Mengurangi beban layanan kesehatan melalui tindakan preventif berbasis data.

##Exploratory Data Analysis

###Import Libraries

```python
# Load libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

###Load Dataset

```python
# Load Dataset
df = pd.read_csv('/content/predic_tabel.csv')
df.head()
```

###Data Preperation

```python
# Data preparation and Exploration
print('Number of Rows: ', df.shape[0])
print('Number of Columns: ', df.shape[1], '\n')
print('SubSet of Data:\n ', df.head(), '\n')
```

###Labeling

```python
# labeling data
labels = ['No',	'Usia',	'Jenis_Kelamin',	'Merokok',	'Bekerja',	'Rumah_Tangga',	'Aktivitas_Begadang',	'Aktivitas_Olahraga',	'Asuransi',	'Penyakit_Bawaan',	'Target']
df.columns = labels
print('Columns Names:', df.columns, '\n')
print('Data Describe:\n ', df.describe(), '\n')
print('Data Information:'); print(df.info())
```

```python
# Convert numerical columns where applicable
for col in ['Usia', 'Jenis_Kelamin', 'Merokok', 'Bekerja', 'Rumah_Tangga',
            'Aktivitas_Begadang', 'Aktivitas_Olahraga', 'Asuransi', 'Penyakit_Bawaan', 'Target']:
    df[col] = df[col].astype('category')
```

###Cek Data yang Duplikat

```python
# Check For duplicated
print(df.duplicated().any())
duplicated = df.duplicated()
print('Number of duplicated data: ', duplicated[duplicated == True].size)
```

###EDA

```python

# Countplot: Distribusi jumlah per kategori pada atribut seperti Merokok, Jenis Kelamin, atau Asuransi
for col in ['Merokok', 'Jenis_Kelamin', 'Asuransi']:
    plt.figure(figsize=(8, 4))
    sns.countplot(data=data, x=col, palette='pastel', edgecolor='black')
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    plt.show()
```

```python
# Heatmap Crosstab: Gambarkan hubungan antara Merokok dan Hasil
plt.figure(figsize=(8, 6))
crosstab = pd.crosstab(data['Merokok'], data['Hasil'])
sns.heatmap(crosstab, annot=True, fmt='d', cmap='YlGnBu')
plt.title('Crosstab Heatmap: Merokok vs Hasil')
plt.xlabel('Hasil')
plt.ylabel('Merokok')
plt.show()
```

```python
# Stacked Bar Chart: Hubungan antara Aktivitas_Olahraga dan Hasil
activity_crosstab = pd.crosstab(data['Aktivitas_Olahraga'], data['Hasil'], normalize='index')
activity_crosstab.plot(kind='bar', stacked=True, figsize=(10, 6), colormap='viridis', edgecolor='black')
plt.title('Stacked Bar Chart: Aktivitas_Olahraga vs Hasil')
plt.xlabel('Aktivitas_Olahraga')
plt.ylabel('Proportion')
plt.legend(title='Hasil')
plt.xticks(rotation=45)
plt.show()
```

```python
# Pie Chart: Persentase kategori dalam Jenis Kelamin
plt.figure(figsize=(6, 6))
data['Jenis_Kelamin'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['lightblue', 'lightgreen'], wedgeprops={'edgecolor': 'black'})
plt.title('Pie Chart: Jenis Kelamin Distribution')
plt.ylabel('')
plt.show()
```
